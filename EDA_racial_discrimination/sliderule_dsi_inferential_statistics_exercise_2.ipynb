I would apply a two-sample t-test. The central limit theorem states that when there is a large number of independent sample data, the distribution is normal even if the population that the data is sampled from is not normal. This data is large (N > 130) and independent therefore, according to the central limit theorem, this data is normally distributed

3)<b>Compute margin of error, confidence interval, and p-value. Try using both the bootstrapping and the frequentist statistical approaches.</b>

```python
w = data[data.race=='w']
b = data[data.race=='b']
#white_data
white_mean = w['call'].mean()
white_std = white_mean*(1-white_mean)
white_N = w['call'].count()
#black data
print('Using the raw data')
black_mean = b['call'].mean()
black_std = black_mean*(1-black_mean)
black_N = b['call'].count()
print('white mean,std,N = {:.4f}, {:.4f}, {}'.format(white_mean,white_std,white_N))
print('black mean,std,N = {:.4f}, {:.4f}, {}'.format(black_mean,black_std,black_N))
# other data
diff_mean = white_mean - black_mean
sum_mean = np.sqrt((white_mean*white_N+black_mean*black_N)/(white_N+black_N))
standard_error = np.sqrt(sum_mean*(1-sum_mean)*(1/white_N+1/black_N))
z_critical = 1.96
margin_of_error = z_critical*standard_error
print('At the 95% confidence interval difference margin of error = {:.4f}'.format(margin_of_error))
print('At the 95% confidence interval of the average difference =[{:.4f},{:.4f}]'
      .format(diff_mean-margin_of_error,diff_mean+margin_of_error))
#z-test
z_score = (white_mean-black_mean)/(standard_error)
pvalue = stats.norm.sf(abs(z_score))
print('z-score, p-value = {:.2f}, {:.2e}'.format(z_score,pvalue))
alpha = 0.05
if diff_mean > 0:
    if pvalue >= alpha:
        print('The null hypothesis that whites and blacks have the same callback rate cannot be reject.')
    else:
        print('The null hypothesis is reject. Whites have a greater callback rate than blacks.')
        #how more often are whites called back.
        dev_mean = white_mean/black_mean
        dev_std = np.sqrt((white_std/white_mean)**2/white_N+(black_std/black_mean)**2/black_N)*dev_mean
        print('Whites get called back {:.2f}(+/-{:.2f}) times as often as blacks'.format(dev_mean,dev_std))    
else:
    print('Blacks have a great callback rate then whites. This is inconsistant with either hypothesis')

#two sample bootstrapping
print('\nUsing bootstrap resampling')
white_bootstrapped = np.empty(10000)
black_bootstrapped = np.empty(10000)
for i in range(10000):
    white_bootstrapped[i] = np.mean(np.random.choice(w['call'],replace=True,size=white_N))
    black_bootstrapped[i] = np.mean(np.random.choice(b['call'],replace=True,size=black_N))
#white data
white_mean = white_bootstrapped.mean()
white_std = white_mean*(1-white_mean)
print('white mean,std = {:.4f}, {:.4f}'.format(white_mean,white_std))
#black data
black_mean = black_bootstrapped.mean()
black_std = black_std*(1-black_std)
print('black mean,std = {:.4f}, {:.4f}'.format(black_mean,black_std))
# other data
diff_mean = white_mean - black_mean
sum_mean = np.sqrt((white_mean*white_N+black_mean*black_N)/(white_N+black_N))
standard_error = np.sqrt(sum_mean*(1-sum_mean)*(1/white_N+1/black_N))
z_critical = 1.96
margin_of_error = z_critical*standard_error
print('At the 95% confidence interval difference margin of error = {:.4f}'.format(margin_of_error))
print('At the 95% confidence interval of the average difference =[{:.4f},{:.4f}]'
      .format(diff_mean-margin_of_error,diff_mean+margin_of_error))
#z-test
z_score = (white_mean-black_mean)/(standard_error)
pvalue = stats.norm.sf(abs(z_score))
print('z-score, p-value = {:.2f}, {:.2e}'.format(z_score,pvalue))
alpha = 0.05
if diff_mean > 0:
    if pvalue >= alpha:
        print('The null hypothesis that whites and blacks have the same callback rate cannot be reject.')
    else:
        print('The null hypothesis is reject. Whites have a greater callback rate than blacks.')
        #how more often are whites called back.
        dev_mean = white_mean/black_mean
        dev_std = np.sqrt((white_std/white_mean)**2/white_N+(black_std/black_mean)**2/black_N)*dev_mean
        print('Whites get called back {:.2f}(+/-{:.2f}) times as often as blacks'.format(dev_mean,dev_std))    
else:
    print('Blacks have a great callback rate then whites. This is inconsistant with either hypothesis')

```
